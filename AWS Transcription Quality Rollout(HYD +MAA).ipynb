```python
## Test Query

sql=f'''
select
b.trans_id,
max(b.'{d1}-{m1}-{y1}') as '{d1}-{m1}-{y1}',
max(b.'{d2}-{m1}-{y1}') as '{d2}-{m1}-{y1}',
max(b.'{d3}-{m1}-{y1}') as '{d3}-{m1}-{y1}',
max(b.'{d4}-{m1}-{y1}') as '{d4}-{m1}-{y1}',
max(b.'{d5}-{m1}-{y1}') as '{d5}-{m1}-{y1}'

from (
select
a.trans_id,
case when a.date='{d1}-{m1}-{y1}' then a.quality else 0 end as '{d1}-{m1}-{y1}',
case when a.date='{d2}-{m1}-{y1}' then a.quality else 0 end as '{d2}-{m1}-{y1}',
case when a.date='{d3}-{m1}-{y1}' then a.quality else 0 end as '{d3}-{m1}-{y1}',
case when a.date='{d4}-{m1}-{y1}' then a.quality else 0 end as '{d4}-{m1}-{y1}',
case when a.date='{d5}-{m1}-{y1}' then a.quality else 0 end as '{d5}-{m1}-{y1}'
from (
select
s.date,
s.trans_id,
s.trans_site,
sum(s.total_defects) as total_defects,
sum(s.total_opportunities) as total_opportunities,
round(100*(1-sum(s.total_defects)/sum(s.total_opportunities)),2) as Quality
from aws_WoW_quality_table s
where trans_id!='none'
group by 1,2
order by 1,2 asc) a
group by 1,a.date
)b
group by 1
''';

    
df_quality_sql = pd.read_sql_query(sql,conn)
df_quality_sql.head(5)
```
