```python
table_opportunties= pd.pivot_table(df, values=['Total Opportunities'],
                         index=['user ID'], 
                         columns= [] ,margins=False,
                         margins_name='total', aggfunc=sum , fill_value=0)
pd.set_option('display.max_rows', df.shape[0]+1)
table_opportunties

table_defects= pd.pivot_table(df, values=['Total Defects'],
                         index=['user ID'], 
                         columns= [] ,margins=False,
                         margins_name='total', aggfunc=sum , fill_value=0)
pd.set_option('display.max_rows', df.shape[0]+1)
table_defects
```

```python
df_SQL = df.rename(columns = {'week': 'week',
                          'Date': 'Date',
                           'user ID' :'userID',
                            'Workflow' :'Work',
                             'file size':'filesize',
                              'TC':'TC',
                              'VC':'VC',
                          'error categorization': 'errorcategorization',
                           'sub categorization': 'subcategorization','Total Opportunities': 'TotalOpportunities',
                          'Total Defects': 'TotalDefects',}, inplace = False)
df_SQL.head()
```

```python
## Weekly Quality Query

sql='''

Select 
Week,
userid,
sum(totalopportunities) as opportunities,
sum(totaldefects) as defects,100*(round(1 - sum(totaldefects)/sum(totalopportunities),4)) as Quality

from aws_verification_quality_summary_db_table
where userID !='NA' and Week=43
group by week,userid

''';

    
df_week_sql = pd.read_sql_query(sql,conn)
df_week_sql.head(100)
```

```python
## Weekly Quality Data wise split Quality Query

sql='''

Select 
Week,date,
userid,
sum(totalopportunities) as opportunities,
sum(totaldefects) as defects,100*(round(1 - sum(totaldefects)/sum(totalopportunities),4)) as Quality

from aws_verification_quality_summary_db_table
where userID !='NA' and Week=43
group by week,date,userid

''';

    
df_date_wise_sql = pd.read_sql_query(sql,conn)
df_date_wise_sql.head(100)
```
