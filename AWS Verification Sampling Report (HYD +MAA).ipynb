```python
## for loop to map the TMs

for i in range (0,len(df)):
    
    if df['Trans ID'].values[i]== '#NA':
        df['Trans_TM'].values[i]='No Match'
    
    if df['Verif_Site'].values[i]=='MAA':
        df['Verif_TM'].values[i]='czvij'
        
    if df['Verif_Site'].values[i]=='HYD':
        df['Verif_TM'].values[i]='latikat'
    
    if df['Trans_Site'].values[i]=='MAA':
        df['Trans_TM'].values[i]='sarselv'
        
    if df['Trans_Site'].values[i]=='HYD':
        df['Trans_TM'].values[i]='latikat'
        
    
                 
        
```

```python
##view pivot
table_sss = pd.pivot_table(df, values=['EC Count'], index=['User ID','Verif_TM'], columns= ['Date'] ,margins=False,
               margins_name='total', aggfunc=sum , fill_value=0)
## display all options
pd.set_option('display.max_rows', df.shape[0]+1)

table_sss.style.set_table_styles(
[{'selector': 'th',
  'props': [('background', '#F5F5F5'), 
            ('color', 'black'),
            ('font-family', 'verdana'),
            ('border-style','solid'),
            ('border-width','thin'),
           ('font-weight', 'normal'),
           ('text-align','center')]},
 
 {'selector': 'td',
  'props': [('font-family', 'verdana'),
           ('border-style','solid'),
            ('border-width','thin'),
           ('text-align','center')]},

 {'selector': 'tr:nth-of-type(odd)',
  'props': [('background', 'white')]}, 
 
 {'selector': 'tr:nth-of-type(even)',
  'props': [('background', 'white')]},
 
 {'selector': 'tr:hover',
  'props': [('background-color', 'yellow')]}

]
).set_precision(2)
```

```python
##to inidivually spot user ID for mails
table_sss.index[0]
```

```python
##Daily Sampling Mail Report to verifers

import win32com.client as win32
outlook = win32.Dispatch('outlook.application')
for i in range(0,len(table_sss)):
    mail = outlook.CreateItem(0)

    mail.Subject = 'AWS Verification Sampling Report'

   
    mail.To = table_sss.index.get_level_values('User ID')[i] + '@amazon.com'
    mail.CC = table_sss.index.get_level_values('Verif_TM')[i] + '@amazon.com'
    
## i-> position of data in dataframe
    mail.HTMLBody = '''<h3>Hi, Please find your sampled counts for the week.</h3>
                   {}'''.format(table_sss[i:i+1].style.set_table_styles(
[{'selector': 'th',
  'props': [('background', '#F5F5F5'), 
            ('color', 'black'),
            ('font-family', 'verdana'),
            ('border-style','double'),
            ('border-width','thin'),
           ('font-weight', 'normal'),
           ('text-align','center')]},
 
 {'selector': 'td',
  'props': [('font-family', 'verdana'),
           ('border-style','double'),
            ('border-width','thin'),
           ('text-align','center'),
           ('text-align','center')]},

 {'selector': 'tr:nth-of-type(odd)',
  'props': [('background', 'white')]}, 
 
 {'selector': 'tr:nth-of-type(even)',
  'props': [('background', 'white')]},
 
 {'selector': 'tr:hover',
  'props': [('background-color', 'yellow')]}

]
).set_precision(2).render())  ## : i + 1 increment 1 of index 
# To attach a file to the email (optional):
#attachment  = "Path to the attachment"
#mail.Attachments.Add(attachment)

    mail.Send()
```

```python
##view pivot for Scorecard ES score
table = pd.pivot_table(df, values=['Total Opportunities'], index=['User ID'], columns= ['Date'] ,margins=True,
               margins_name='total', aggfunc=np.mean , fill_value=0)
## display all options
pd.set_option('display.max_rows', df.shape[0]+1)
table.round(2)
```

```python
##view pivot for scorecard defects
table_defects = pd.pivot_table(df, values=['Total Defects'], index=['User ID'], columns= ['Date'] ,margins=True,
               margins_name='total', aggfunc=sum , fill_value=0)
## display all options
pd.set_option('display.max_rows', df.shape[0]+1)
table_defects.round(2)
```
