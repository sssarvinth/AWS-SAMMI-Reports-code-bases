```python
df['File compare'] = df['Audio Duration (HH:MM:SS.mss)'].eq(df['Audio Duration (HH:MM:SS.mss)'].shift())

df.head()
```

```python
## change the true or false to 1 or 0, match is true, mismatch is false
for column in df['File compare']:
    df['File compare'] = df['File compare'].replace(True,1)
    df['File compare'] = df['File compare'].replace(False,0) 

#pd.set_option('display.max_rows',None)
df.head()
```

```python
##since all the sheets of each user ID is merged, we need to reset the index value, this will help us map the for loop.
##without resting the index value will revert to 0 once the new user id occurs, with the reset value it will continue to increment
## for example without index 0,1,2,3,0,1,2,3, and so on, with index rest 0,1,2,3,4,..100
##the updated index will map righlty to the completed audio lengths.

df.reset_index(inplace=True)
```
